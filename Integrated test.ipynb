```python
import pandas as pd
import optimizer_module as om
import price_prediction_v2 as pp
import datetime
from math import floor

start_time = datetime.datetime.now() # Start time

# Starting variables
numDays = 1  # Number of days to run battery
intPerOpt = 216 # Number of time intervals optimizer is fed -> 216 = optimizes for 18 hours ahead
start = 0 # Starting time interval for price data
bStorage0 = 0 # Starting battery charge

# Don't Change
intPerDay = 288 # Number of time intervals per day

# Chosen time intervals
times = pd.read_csv('final.csv', parse_dates=True).iloc[start:start+(numDays*intPerDay)]["SETTLEMENTDATE"].tolist()

# Real prices
realPrices = pd.read_csv('final.csv', parse_dates=True).iloc[start:start+(numDays*intPerDay)]["RRP5MIN"].tolist()
realPrices = [i * 12 for i in realPrices] # prices divided by 12


predPrice = pp.price_prediction('final.csv')
print(predPrice)
#predPrices = realPrices # Just to test

maxProfit = 0 # Profit by running optimizer using real prices
actualProfit = 0 # profit by running optimizer using predicted prices

# Starting charge level in battery
realNxtBatCharge = bStorage0
predNxtBatCharge = bStorage0

# Saves actions of battery optimizing for real and predicted prices
realActions = [] 
predActions = [] 

# Saves battery levels of optimal battery use for real and predicted price
realBatCharge = [realNxtBatCharge] 
predBatCharge = [predNxtBatCharge] 

# Saves ongoing profit of optimal battery use for real and predicted price
realProfit = [0] 
predProfit = [0]

# Runs battery for numDays
for i in range(numDays*intPerDay):
    # Prints update when battery has started running for a new day
    if (i % intPerDay == 0):
        print("Running day %g..." % (floor(i/(intPerDay-1))+1))
    
    # Runs optimzer for real prices
    realNxtAction, realNxtBatCharge = om.optimize(realPrices[i:i+intPerOpt], realNxtBatCharge)
    maxProfit += realNxtAction*realPrices[i]/12 # Max profit is calculated by running optimizer using real prices
    realActions.append(realNxtAction)
    realBatCharge.append(realNxtBatCharge)
    realProfit.append(maxProfit)
    
    # predPrices = GET PREDICTED PRICES FOR correct time intervals -> will probably have to change om.optimize input below
    
    # Runs optimzer for predicted prices
    predNxtAction, predNxtBatCharge = om.optimize(predPrices[i:i+intPerOpt], predNxtBatCharge) # get rid of i once price predictor added
    actualProfit += predNxtAction*realPrices[i]/12 # Actual profit by running optimizer using predicted prices
    predActions.append(predNxtAction)
    predBatCharge.append(predNxtBatCharge)
    predProfit.append(actualProfit)
    
    
    # Prints update when battery has run for an entire day
    if (i % intPerDay == intPerDay-1):
        print("Completed day %g of %g" % (floor(i/(intPerDay-1)), numDays))
    
    
elapsed_time = datetime.datetime.now() - start_time

# Prints Results
print("\n--------------------RESULTS--------------------")
print("\nRan battery for %g day/s" % (numDays))
print("Elapsed time: " + str(elapsed_time))
print("Optimal profit: $%.2f" % (maxProfit))
print("Actual profit: $%.2f -> %.2f%% of optimal profit" % (actualProfit,actualProfit/maxProfit*100))   

# Prints battery use for our battery vs optimal use of battery
predBatCharge.pop(), realBatCharge.pop(), predProfit.pop(), realProfit.pop()
results = pd.DataFrame({'Price ($)': realPrices, 'State': predActions, 'Bat Lvl (MW)': predBatCharge,
     'Profit ($)': predProfit,'Optimal State': realActions, 'Optimal Bat Lvl (MW)': realBatCharge,
        'Optimal Profit ($)': realProfit}, index=times)
results
```
